### **6.Loops**

A loop in Python is a programming control structure used to execute a block of code repeatedly. Instead of writing the same line of code multiple times, loops allow you to automate repetitive tasks efficiently.

Python offers two primary types of loops: the for loop and the while loop.

1. The **for** Loop
You use a for loop when you want to iterate over a sequence (like a list, tuple, dictionary, set, or string) or a known range of numbers. It runs a predetermined number of times based on the length of that sequence.

   **When to use:** When you know ahead of time how many times you need to repeat the action.

2. The **while** Loop
A while loop executes a block of code as long as a specific condition remains True.

   **When to use:** When you do not know the exact number of iterations beforehand, and execution depends strictly on a changing state.

**for Loop**

Used to loop over a sequence (list, dictionary, range) when you know the number of iterations in advance.

```python
#For Loop

#Iterating over a list

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
#Using the range() function

# Prints numbers from 0 to 4

for i in range(5):
    print(i)
```

```python
scores = [80, 90, 75]

for score in scores:
    print(score)
```

**while Loop**

Executes a block of code as long as a condition is True.

```python
#while Loop

count = 0
while count < 3:
    print(count)
    count += 1  # Increments count so the loop doesn't run forever
```

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
#Sum Numbers

total = 0

for i in range(1, 6):
    total += i

print(total)
```

The Python range(1, 6) function generates a sequence of numbers starting at 1 and ending right before 6 (numbers 1, 2, 3, 4, 5).

Here is exactly how the total variable changes during each iteration of the loop:

Initial State: total is set to 0.

Loop 1: i is 1. total becomes 0 + 1 = 1.

Loop 2: i is 2. total becomes 1 + 2 = 3.

Loop 3: i is 3. total becomes 3 + 3 = 6.

Loop 4: i is 4. total becomes 6 + 4 = 10.

Loop 5: i is 5. total becomes 10 + 5 = 15.

```python
#Count Participants

participants = ["P1", "P2", "P3", "P4"]

count = 0

for p in participants:
    count += 1

print("Total:", count)
```

### **7. Functions**

A function in Python is a self-contained, reusable block of organized code designed to perform a specific task. Instead of writing the same lines of code repeatedly, you encapsulate that logic into a single named block and call it whenever needed, following the **DRY (Don't Repeat Yourself)** principle.

A reusable block of code that runs only when it is called. You can pass data (parameters) into it and return data back.

```python
#Simple Function

def greet():
    print("Hello")

greet()
```

```python
#Function with Parameter

def greet(name):
    print("Hello", name)

greet("Rahim")
greet("Karim")
```

```python
#Add Numbers

def add(a, b):
    return a + b

result = add(10, 20)

print(result)
```

```python
#Calculate Average

def average(numbers):
    return sum(numbers) / len(numbers)

scores = [80, 85, 90]

print(average(scores))
```

```python
#Grade Function

def grade(marks):

    if marks >= 80:
        return "A+"

    elif marks >= 70:
        return "A"

    else:
        return "B"

print(grade(85))
print(grade(75))
print(grade(60))
```

**Combined Real-World Example**

```python
#Combined Real-World Example

#Suppose you have student marks and want to determine who passed.

students = {
    "Rahim": 85,
    "Karim": 65,
    "Jamal": 45
}

def result(mark):

    if mark >= 50:
        return "Pass"
    else:
        return "Fail"

for name, mark in students.items():

    print(name, mark, result(mark))
```

## **8. Dictionary**

A Python dictionary is a built-in data structure that stores information in key-value pairs. Instead of using a numeric index to look up items like a list, you retrieve data using a descriptive name or "key". You can think of it like a real-world phone book, where you look up a person's name (the key) to find their phone number (the value).

**Essential Dictionary Methods**
* .**get(key)**: Retrieves a value safely without throwing a KeyError if the key doesn't exist.
* **.keys()**: Returns a list-like view of all keys in the dictionary.
* **.values()**: Returns a list-like view of all values.
* **.items()**: Returns key-value pairs as a collection of tuples, ideal for loops

```python
#Creating a Dictionary

student = {
    "name": "Rahim",
    "age": 22,
    "cgpa": 3.85
}

print(student)
```
